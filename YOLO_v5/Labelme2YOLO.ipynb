```python
import os
import json

folder_path = '/content/drive/MyDrive/CMPT898_Project/898_Files/Annotated_Dataset/LR-70_Annotated'

# 遍歷資料夾中的所有檔案
for filename in os.listdir(folder_path):
    if filename.endswith('.json'):
        file_path = os.path.join(folder_path, filename)

        # 讀取 JSON 檔案
        with open(file_path, 'r', encoding='utf-8') as f:
            data = json.load(f)

        # 修改 label 名稱
        for shape in data['shapes']:
            if shape['label'] == 'nodules':
                shape['label'] = 'Nodule'

        # 寫回修改後的 JSON 檔案
        with open(file_path, 'w', encoding='utf-8') as f:
            json.dump(data, f, ensure_ascii=False, indent=4)

```

```python
import os
import shutil

def merge_labelme_annotations(src_dirs, dest_dir):
    os.makedirs(dest_dir, exist_ok=True)
    for src_dir in src_dirs:
        for filename in os.listdir(src_dir):
            src_file = os.path.join(src_dir, filename)
            dest_file = os.path.join(dest_dir, filename)
            shutil.copy(src_file, dest_file)

# Define source directories and destination directory
source_directories = [
    '/content/drive/MyDrive/CMPT898_Project/898_Files/Annotated_Dataset/LR-68_Annotated',
    '/content/drive/MyDrive/CMPT898_Project/898_Files/Annotated_Dataset/LR-70_Annotated',
    '/content/drive/MyDrive/CMPT898_Project/898_Files/Annotated_Dataset/LR-86_Annotated',
    '/content/drive/MyDrive/CMPT898_Project/898_Files/Annotated_Dataset/The_Missing_Annotations'
]
destination_directory = '/content/drive/MyDrive/CMPT898_Project/898_Files/Annotated_Dataset/Combined_Annotations'

# Merge annotations
merge_labelme_annotations(source_directories, destination_directory)
```
